```python
import pandas as pd 
import networkx as nx
```

```python
df=pd.read_csv('train.csv')
```

```python
df
```

```python
df['subject_entity'][0].split(',')[3].split(':')[1].replace('}','').lstrip()
```

```python
df['subject_word'] =df['subject_entity'].apply(lambda x: x.split(',')[2].split(':')[1].lstrip())
df['object_word'] =df['object_entity'].apply(lambda x: x.split(',')[2].split(':')[1].lstrip())
df['subject_type'] =df['subject_entity'].apply(lambda x: x.split(',')[3].split(':')[1].replace('}','').lstrip())
df['object_type'] =df['object_entity'].apply(lambda x: x.split(',')[3].split(':')[1].replace('}','').lstrip())
df['word_node']=df['subject_word']+'-'+df['object_word']
df['type_node']=df['subject_type']+'-'+df['object_type']
df['full_word']=df['word_node']+'^'+df['label']
df['full_type']=df['type_node']+'^'+df['label']
```

```python
from collections import Counter
```

```python
Counter(df['full_word'])
```

```python
word_df = pd.DataFrame.from_dict(Counter(df['full_word']),orient='index',columns=['weight']).reset_index()
type_df = pd.DataFrame.from_dict(Counter(df['full_type']),orient='index',columns=['weight']).reset_index()
```

```python
word_df['Source'] = word_df['index'].apply(lambda x: x.split('^')[0])
word_df['Target'] = word_df['index'].apply(lambda x: x.split('^')[1])
type_df['Source'] = type_df['index'].apply(lambda x: x.split('^')[0])
type_df['Target'] = type_df['index'].apply(lambda x: x.split('^')[1])
```

```python
word_df.sort_values(by='weight',ascending=False)
```

```python
type_df.sort_values(by='weight',ascending=False)
```

```python
word_G = nx.from_pandas_edgelist(word_df,
                                source='Source',
                                target='Target',
                                edge_attr='weight')
```

```python
type_G = nx.from_pandas_edgelist(type_df,
                                source='Source',
                                target='Target',
                                edge_attr='weight')
```
